## KNN

### Siapkan data

```python
nama = ["Udin", "Jenap", "Mamang", "Selametkopling", "MadTopo"]
umur = [22, 33, 25, 22 ,22]
Hobi = ["Berenang", "Berenang", "Main game", "Main game", "Berenang"]
Gender = ["Pria", "Wanita", "Pria", "Pria", "Pria"]
```

```python
data = {
    "Nama":nama,
    "Umur":umur,
    "Gender":Gender,
    "Hobi":Hobi
}
df = pd.DataFrame(data)
```

```python
df
```

```python
def encoder_hoby(data):
  hobi ={
    "Berenang": 1,
    "Main game": 0
  }
  return hobi.get(data)

def encoder_gender(data):
  gender ={
    "Pria": 1,
    "Wanita": 0
  }
  return gender.get(data)
```

```python
df["Hobi_transform"] = df["Hobi"].apply(encoder_hoby)
df["Gender_transform"] = df["Gender"].apply(encoder_gender)
```

```python
data_prep = df[["Umur","Gender_transform","Hobi_transform"]]
```

```python
data_prep
```

```python
x1 = data_prep["Umur"]
x2 = data_prep["Gender_transform"]
```

### Prediksi

Prediction
* Sarah gender Wanita umur 17 mempunyai Whats Hobi ?

```python
umur_pred = 17
gender_pred = 0 #wanita
```

```python
hasil_distance = []
for umur, gender in zip(x1, x2):
  ec = np.sqrt( np.power([umur_pred - umur], 2) + np.power([gender_pred - gender], 2))
  hasil_distance.append( ec.tolist() )
```

```python
import itertools
rs = list(itertools.chain(*hasil_distance))
distance = pd.DataFrame({
    "distance":rs
})
data = pd.concat([data_prep,distance],axis=1)
```
